```python
def get_stats(data,col_list,target):

    main_stats = pd.DataFrame(columns = ['Feature','Levels','Class count','Event count','Non-Event count'])
    data_change = data
    for col in col_list:

        data_filter = data.dropna(subset = [col])
        elements = data_filter[col].unique().tolist()

        for element in elements:
            stats ={}

            stats['Feature'] = col
            stats['Levels'] = element
            stats['Class count'] = data_filter[data_filter[col]==element].shape[0]
            stats['Event count'] = data_filter[data_filter[col]==element].loc[data_filter[target]==1,].shape[0]
            stats['Non-Event count'] = data_filter[data_filter[col]==element].loc[data_filter[target]==0,].shape[0]

            #ClassCount.append(data[data[col]==element].shape[0])
            #Event_Count.append(data[data[col]==element][data['target']==1].shape[0])
            #Non_Event_Count.append(data[data[col]==element][data['target']==0].shape[0])

            main_stats =main_stats.append(stats,ignore_index = True)

    return main_stats
```
